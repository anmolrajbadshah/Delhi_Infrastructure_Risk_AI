```python
from pyrosm import OSM
import geopandas as gpd
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (8,8)
```

```python
# Delhi NCR approx bounding box
# (west, south, east, north)
delhi_ncr_bbox = [76.5, 28.2, 77.7, 29.0]
```

```python
osm = OSM(
    "gis/raw/northern-zone-260116.osm.pbf",
    bounding_box=delhi_ncr_bbox
)




```

```python
boundaries = osm.get_boundaries()
boundaries.head()
```

```python
boundaries["admin_level"].value_counts()
```

```python
delhi_ncr = boundaries[
    (boundaries["admin_level"] == 6) &
    (boundaries["name"].str.contains(
        "Delhi|New Delhi|Central Delhi|North Delhi|South Delhi|East Delhi|West Delhi|"
        "Ghaziabad|Noida|Gurgaon|Gurugram|Faridabad",
        case=False,
        na=False
    ))
]
```

```python
#ax = delhi_ncr.plot(edgecolor="black", figsize=(8,8))
#ax.set_aspect("auto")
```

```python
delhi_ncr = delhi_ncr.to_crs(epsg=4326)

delhi_ncr.to_file(
    "gis/processed/delhi_ncr_districts.geojson",
    driver="GeoJSON"
)
```
